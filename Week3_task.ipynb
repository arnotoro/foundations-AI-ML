```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# Variable for infinity
inf = np.inf

# Adjacency cost matrix from exercise 
adj_matrix = np.array([
    [inf, 5, inf, inf, inf],
    [5, inf, 7, 10, inf],
    [inf, 7, inf, 2, inf],
    [inf, 10, 2, inf, 6],
    [inf, inf, inf, 6, inf]
])

# Inf as 0 for networkx
adj_matrix[adj_matrix == inf] = 0

# Create a graph from the adjacency cost matrix
G = nx.from_numpy_array(adj_matrix)

# Draw the graph with edge labels
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=15)

# Draw the graph
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

# Show the graph
plt.show()
```
